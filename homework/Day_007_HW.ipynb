```python
import os
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
# 設定 data_path, 並讀取 app_train
dir_data = 'C:/Users/user/Desktop/Python'
f_app_train = os.path.join(dir_data, 'application_train.csv')
df1 = pd.read_csv(f_app_train)
```

```python
###
### Day 7
###
day7_1 =df1['AMT_ANNUITY']
print(day7_1.head())

###
# 1 列出 AMT_ANNUITY 的 q0 - q100
###

q0 = min(day7_1)
q100 = max(day7_1)
print(q0)
print(q100)

###
# 2.1 將 AMT_ANNUITY 中的 NAs 暫時以中位數填補
###

#如果有NA，describe會有問題
day7_1.describe()

# Ignore NA, 計算五值
five_num = [0, 25, 50, 75, 100]
quantile_5s = [np.percentile(df1[~day7_1.isnull()]['AMT_ANNUITY'], q = i) for i in five_num]
#print(quantile_5s)

#畫出hist plot
df1[~day7_1.isnull()]['AMT_ANNUITY'].hist(bins = 100)
plt.show()

#計算每行NA數量
missing_values_count = df1.isnull().sum()
print(missing_values_count[:10])

#删除数据表中含有空值的行
#df1.dropna(how='any')

df1['AMT_ANNUITY'].fillna(df1['AMT_ANNUITY'].median())

#以平均數填補
#df1['AMT_ANNUITY'].fillna(df1['AMT_ANNUITY'].mean())

###
# 2.2 將 AMT_ANNUITY 的數值標準化至 -1 ~ 1 間
###

#Original
day7_1.hist(bins = 50)
plt.title("Original")
plt.show()
value = day7_1
#value = df1['AMT_ANNUITY'].values

#Z-transform
df1['AMT_ANNUITY_Norm1'] = ( value - np.mean(value) ) / ( np.std(value) )
df1['AMT_ANNUITY_Norm1'].hist(bins = 50)
plt.title("Normalized with Z-transform")
plt.show()

#Normalized to 0 ~ 1
df1['AMT_ANNUITY_Norm2'] = ( value - min(value) ) / ( max(value) - min(value) )
df1['AMT_ANNUITY_Norm2'].hist(bins = 50)
plt.title("Normalized to 0 ~ 1")
plt.show()
#3   將 AMT_GOOD_PRICE 的 NAs 以眾數填補
day7_3 = df1['AMT_GOODS_PRICE']
day7_3.fillna(day7_3.mode())
```
